#### 1. Solving a System of Linear Equations

##### Use Python to solve the following system of equations. Hint: You can write this system as a matrix equation in in the form Ax = b, where the matrix A is invertible

```python
import numpy as np
import matplotlib.pyplot as plt

#create matrix
a = np.matrix([[27, -10, 4, -29], [-16, 5, -2, 18], [-17, 4, -2, 20], [-7, 2, -1, 8]])
b = np.matrix([[1], [-1], [0], [1]])

#Solve the system of equations
x = np.linalg.solve(a, b)

#calcula A*X, and allocate a new data type (int)
ax = (a*x).astype(int)

#display the results
print("A =\n", a)
print("\nb =\n", b)
print("\nX =\n", x)
print("\nAX =\n", ax)

```

#### 2. Visualizing a Linear Transformation

##### Let T be the linear transformation that takes in a vector v in R2 and outputs the product of the 2Ã—2 matrix A shown below and v.
##### I.e., T is a function that takes in a vector v in R2 and outputs the vector T(v)= Av, also in R2.

```python
A = np.matrix([[np.sqrt(2)/2, -np.sqrt(2)/2], [np.sqrt(2)/2, np.sqrt(2)/2]])
A
```

##### b) Mimic the code in Section 7.1.3 of Essential Math for Data Science to visualize the linear transformation T.

```python
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)

xx, yy = np.meshgrid(x, y)
plt.scatter(xx, yy, s=20, c=xx+yy)
plt.show()
```

```python
xy = np.vstack([xx.flatten(), yy.flatten()])
xy.shape
```

```python
trans = np.array(A@xy)
trans.shape
```

```python
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```
